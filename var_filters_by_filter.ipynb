```python
# redo variant level filters to be all chromosomes one filter per file
import pandas as pd
```

```python
# columns
df = pd.read_csv('variant_filters_by_chrom/variant_filters_chr22.tsv', sep='\t')
columns = df.columns
```

```python
columns
```

```python
filters = ['mu_decile_0', 'mu_decile_1', 'mu_decile_2', 'mu_decile_3', 'mu_decile_4',
           'mu_decile_5', 'mu_decile_6', 'mu_decile_7', 'mu_decile_8', 'mu_decile_9', 
           'predicted_pocket', 'nonpolar_neutral_aa', 'polar_neutral_aa', 'charged_aa', 
           'aromatic_aa', 'destabilizing', 'non_destabilizing', 'disordered', 
           'ordered', 'buried', 'surface']
```

```python
final_columns = ['chrom', 'pos', 'ref', 'alt', 'ensg', 'enst', 'uniprot_isoform', 'aa_pos', 'aa_ref', 'aa_alt']
```

```python
chromosomes = [(str(i)) for i in range(1, 23)] + ['X']
```

```python
for f in filters:
    print(f)
    df_big = pd.DataFrame()
    for c in chromosomes:
        print(c)
        df = pd.read_csv(f'variant_filters_by_chrom/variant_filters_chr{c}.tsv', 
                         sep='\t', usecols=final_columns+[f])
        df = df[df[f]==True]
        df_big = pd.concat([df_big, df])
    df_big = df_big.reset_index(drop=True)
    df_big = df_big[final_columns]
    print(f'saving filter {f}')
    df_big.to_csv(f'variant_filters_by_filter/variant_filters_{f}.tsv.gz', sep='\t', compression='gzip', index=False)
```

```python
# fix cath_class
f = 'cath_class'
classes = [1,2,3,4,6]
for cl in classes:
    cl_i = int(cl)
    print(cl_i)
    df_big = pd.DataFrame()
    for c in chromosomes:
        print(c)
        df = pd.read_csv(f'variant_filters_by_chrom/variant_filters_chr{c}.tsv', 
                         sep='\t', usecols=final_columns+[f])
        df = df[df[f]==cl_i]
        df_big = pd.concat([df_big, df])
    df_big = df_big.reset_index(drop=True)
    df_big = df_big[final_columns]
    print(f'saving class {cl_i}')
    df_big.to_csv(f'variant_filters_by_filter/variant_filters_cath_class_{cl_i}.tsv.gz', sep='\t', compression='gzip', index=False)
```

```python
# get list of all architectures
archs = pd.read_csv('../domains/uniprot_architecture_clean.tsv', sep='\t', dtype={'CATH_architecture':str})
a_list = archs.CATH_architecture.unique()
```

```python
a_list
```

```python
# fix 'cath_architecture'
for a in a_list:
    print(a)
    df_big = pd.DataFrame()
    for c in chromosomes:
        print(c)
        df = pd.read_csv(f'variant_filters_by_chrom/variant_filters_chr{c}.tsv', 
                         sep='\t', usecols=final_columns+['gene3d_id'])
        df['cath_arch'] = df.gene3d_id.str.split('G3DSA:').str[1].str.split('.').str[0]+'.'+df.gene3d_id.str.split('G3DSA:').str[1].str.split('.').str[1]
        df = df[df['cath_arch']==a]
        df = df[final_columns]
        df_big = pd.concat([df_big, df], ignore_index=True)
    print(f'saving class {a}')
    df_big.to_csv(f'variant_filters_by_filter/variant_filters_cath_arch_{a.replace('.', '_')}.tsv.gz', sep='\t', compression='gzip', index=False)
```
